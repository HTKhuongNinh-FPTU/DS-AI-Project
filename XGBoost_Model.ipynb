```python
features = [
    'store_encoded', 'product_encoded','discount', 'holiday_flag', 'activity_flag',
    'precpt', 'avg_temperature', 'avg_humidity', 'avg_wind_level']

X = df[features]
Y = y_hours_df
df_sorted = df.sort_values(by='dt').reset_index(drop=True)
Y_sorted = Y.loc[df_sorted.index]
unique_dates = df_sorted['dt'].dt.date.unique()
num_unique_dates = len(unique_dates)
train_days = 83
test_days = 7
if num_unique_dates < (train_days + test_days):
    print(f"Warning: Not enough unique dates for the requested split. Found {num_unique_dates} dates, need at least {train_days + test_days}.")
    train_cutoff_date = unique_dates[min(train_days - 1, num_unique_dates - test_days - 1)]
else:
    train_cutoff_date = unique_dates[train_days - 1]
train_indices = df_sorted[df_sorted['dt'].dt.date <= train_cutoff_date].index
test_indices = df_sorted[df_sorted['dt'].dt.date > train_cutoff_date].index

X_train = X.iloc[train_indices]
X_test = X.iloc[test_indices]
Y_train = Y.iloc[train_indices]
Y_test = Y.iloc[test_indices]

print(f"Kích thước tập huấn luyện X_train: {X_train.shape}")
print(f"Kích thước tập kiểm thử X_test: {X_test.shape}")
print(f"Số ngày huấn luyện: {len(df_sorted.loc[train_indices, 'dt'].dt.date.unique())}")
print(f"Số ngày kiểm thử: {len(df_sorted.loc[test_indices, 'dt'].dt.date.unique())}")
```

```python
# Khởi tạo lõi phân loại XGBoost
xgb_core = xgb.XGBClassifier(
    n_estimators=50,
    max_depth=5,
    learning_rate=0.1,
    random_state=42,
    eval_metric='logloss')
# Sử dụng MultiOutputClassifier để nhân bản lõi XGBoost dự đoán song song cho 16 khung giờ
multi_target_model = MultiOutputClassifier(xgb_core, n_jobs=-1)
multi_target_model.fit(X_train, Y_train)
print("Mô hình đã học xong quy luật vận hành của 16 tiếng hoạt động.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import pandas as pd
# Danh sách 16 ngưỡng tối ưu bạn vừa tìm được
optimized_thresholds = [
    0.196, 0.219, 0.237, 0.233, 0.175, 0.196, 0.205, 0.206,
    0.219, 0.239, 0.251, 0.279, 0.299, 0.328, 0.337, 0.345
]

y_pred_proba_list = multi_target_model.predict_proba(X_test)
# Thực hiện dự báo với ngưỡng tùy biến cho từng giờ
Y_pred_thresholded = np.zeros((X_test.shape[0], len(operating_hours_names)))

for i, proba_hour in enumerate(y_pred_proba_list):
    # Dùng ngưỡng tương ứng với khung giờ i
    Y_pred_thresholded[:, i] = (proba_hour[:, 1] >= optimized_thresholds[i]).astype(int)

# Chuyển thành DataFrame để tính toán chỉ số
Y_pred_df = pd.DataFrame(Y_pred_thresholded, columns=operating_hours_names)
Y_pred = Y_pred_thresholded
acc_list = []
precision_list = []
recall_list = []
f1_list = []
for col in operating_hours_names:
    y_true_hour = Y_test[col]
    y_pred_hour = Y_pred_df[col]
    acc_list.append(accuracy_score(y_true_hour, y_pred_hour))
    precision_list.append(precision_score(y_true_hour, y_pred_hour, zero_division=0))
    recall_list.append(recall_score(y_true_hour, y_pred_hour, zero_division=0))
    f1_list.append(f1_score(y_true_hour, y_pred_hour, zero_division=0))
acc = np.mean(acc_list)
precision = np.mean(precision_list)
recall = np.mean(recall_list)
f1 = np.mean(f1_list)
print("\n========= KẾT QUẢ KIỂM ĐỊNH MÔ HÌNH 16 TIẾNG HOẠT ĐỘNG ==========")
print(f"Tổng số ca kiểm tra (Dòng dữ liệu tập Test): {len(X_test)}")
print(f"1. Độ chính xác tổng thể (Accuracy): {acc:.2%}")
print(f"2. Độ chính xác dự báo đứt hàng (Precision): {precision:.2%}")
print(f"3. Khả năng tìm thấy ca đứt hàng (Recall): {recall:.2%}")
print(f"4. Điểm cân bằng (F1-Score): {f1:.2%}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import recall_score
import numpy as np
sns.set_theme(style="white")
recall_by_hours = []
missed_by_hours = []
for col in operating_hours_names:
    y_true_hour = Y_test[col]
    y_pred_hour = Y_pred_df[col]

    hour_recall = recall_score(y_true_hour, y_pred_hour, zero_division=0)
    recall_by_hours.append(hour_recall)

    missed_by_hours.append(1.0 - hour_recall)
hours = range(6, 22)
plt.figure(figsize=(15, 6.5))
bars_captured = plt.bar(hours, recall_by_hours, label='Tìm thấy (Đoán trúng)', color='#2ca02c', alpha=0.85, edgecolor='black', linewidth=0.8)
bars_missed = plt.bar(hours, missed_by_hours, bottom=recall_by_hours, label='Bỏ sót (Đoán sai)', color='#ff9896', alpha=0.7, edgecolor='black', linewidth=0.8)

for i in range(len(hours)):
    captured_val = recall_by_hours[i]
    missed_val = missed_by_hours[i]
    if captured_val > 0.05:
        plt.text(hours[i], captured_val / 2, f"{captured_val:.1%}", ha='center', va='center', color='white', fontweight='bold', fontsize=9)
    if missed_val > 0.05:
        plt.text(hours[i], captured_val + (missed_val / 2), f"{missed_val:.1%}", ha='center', va='center', color='black', fontweight='bold', fontsize=9)
plt.title('PHÂN TÍCH TỶ LỆ TÌM THẤY VÀ BỎ SÓT TRÊN TỔNG SỐ CA HẾT HÀNG THỰC TẾ (16 TIẾNG)', fontsize=13, fontweight='bold', pad=20)
plt.xlabel('Khung giờ hoạt động thực tế của siêu thị (Từ 6h đến 21h)', fontsize=11, labelpad=10)
plt.ylabel('Tỷ lệ đóng góp trên 100% ca thực tế đứt hàng', fontsize=11, labelpad=10)
plt.xticks(hours)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
plt.ylim(0, 1.05)
plt.legend(fontsize=10, loc='upper right', bbox_to_anchor=(1, 1.03), ncol=2, frameon=True)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
